```python
# Split the genres string and create a list of all genres
genres_list = [genre.strip("[]").replace("'", "").split(", ") for genre in df['Genres']]
all_genres = {genre for sublist in genres_list for genre in sublist}
all_genres.remove('')

print(all_genres)
```
